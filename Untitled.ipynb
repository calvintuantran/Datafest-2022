```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
edit_database = pd.read_csv("use.csv", low_memory=False)
edit_database
```

```python
import seaborn as sns
sns.set_theme()

# Plot sepal width as a function of sepal_length across days
g = sns.lmplot(
    data=edit_database,
    x="age", y="max_progression", hue="School",
    height=5
)

# Use more informative axis labels than are provided by default
g.set_axis_labels("age", "max_progression")
```

```python
import seaborn as sns
sns.set_theme(style="ticks")

# Load the example dataset for Anscombe's quartet
df = sns.load_dataset("anscombe")

# Show the results of a linear regression within each dataset
sns.lmplot(x="age", y="max_progression", col="ethnicity", hue="ethnicity", data=edit_database,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```
